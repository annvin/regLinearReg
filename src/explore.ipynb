```python
# Your code here
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
main_df = pd.read_csv('https://raw.githubusercontent.com/4GeeksAcademy/regularized-linear-regression-project-tutorial/main/demographic_health_data.csv')
```

```python
main_df
```

```python
drop_list = ['COUNTY_NAME','STATE_NAME','fips']
```

```python
main_df = main_df.drop(columns=drop_list)
```

```python
main_df
```

```python
main_df.columns.tolist()
```

```python
drop_list_1 = ['Heart disease_prevalence',
 'Heart disease_Lower 95% CI',
 'Heart disease_Upper 95% CI']
```

```python
main_df = main_df.drop(columns=drop_list_1)
```

```python
y = main_df['Heart disease_number']
```

```python
X = main_df.drop(columns=['Heart disease_number'])
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, random_state=4)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
linreg = LinearRegression()
```

```python
linreg.fit(X_train, y_train)
```

```python
train_preds = linreg.predict(X_train)
```

```python
test_preds = linreg.predict(X_test)
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error 
```

```python
print(f'training set r2 score {r2_score(y_train, train_preds)}')
```

```python
print(f'training set mean absolute error {mean_absolute_error(y_train, train_preds)}')
```

```python
print(f'training set mean squared error {np.sqrt(mean_squared_error(y_train, train_preds))}')
```

```python
print(f'testing set r2 score {r2_score(y_test, test_preds)}')
print(f'testing set mean absolute error {mean_absolute_error(y_test, test_preds)}')
print(f'testing set mean squared error {np.sqrt(mean_squared_error(y_test, test_preds))}')
```

```python
sns.scatterplot(x = y_train, y=train_preds)
```

```python
sns.scatterplot(x = y_test, y=test_preds)
```

```python
from sklearn.linear_model import Lasso
```

```python

lasso_model = Lasso(alpha= 10, max_iter= 5000)

```

```python
lasso_model.fit(X_train, y_train)
```

```python
lasso_train_preds = lasso_model.predict(X_train)
```

```python
lasso_test_preds = lasso_model.predict(X_test)
```

```python
print(f'testing set r2 score {r2_score(y_train, lasso_train_preds)}')
print(f'testing set mean absolute error {mean_absolute_error(y_train, lasso_train_preds)}')
print(f'testing set mean squared error {np.sqrt(mean_squared_error(y_train, lasso_train_preds))}')
```

```python
print(f'testing set r2 score {r2_score(y_test, lasso_test_preds)}')
print(f'testing set mean absolute error {mean_absolute_error(y_test, lasso_test_preds)}')
print(f'testing set mean squared error {np.sqrt(mean_squared_error(y_test, lasso_test_preds))}')
```

```python
hyperparameters = {'alpha':[0,0.1,1,10,100],
'max_iter':[50,100,1000,2000,5000]}

from sklearn.model_selection import GridSearchCV

grid_model = GridSearchCV(Lasso(), param_grid=hyperparameters, scoring = 'neg_root_mean_squared_error' )

grid_model.fit(X_train, y_train)

```

```python
grid_model.best_params_
```

```python
from sklearn.linear_model import Ridge
```

```python

ridge_model = Ridge(alpha=100, max_iter=50)
```

```python
ridge_model.fit(X_train, y_train)
```

```python
ridge_train_preds = ridge_model.predict(X_train)
```

```python
ridge_test_preds = ridge_model.predict(X_test)
```

```python
print(f'testing set r2 score {r2_score(y_train, ridge_train_preds)}')
print(f'testing set mean absolute error {mean_absolute_error(y_train, ridge_train_preds)}')
print(f'testing set mean squared error {np.sqrt(mean_squared_error(y_train, ridge_train_preds))}')
```

```python
print(f'testing set r2 score {r2_score(y_test, ridge_test_preds)}')
print(f'testing set mean absolute error {mean_absolute_error(y_test, ridge_test_preds)}')
print(f'testing set mean squared error {np.sqrt(mean_squared_error(y_test, ridge_test_preds))}')
```

```python
lasso_grid_results_df = pd.DataFrame(grid_model.cv_results_)
```
